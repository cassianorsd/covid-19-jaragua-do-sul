```python
import pandas as pd
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_pickle('./cases.pkl')
```

```python
df = df.sort_values(by='date',ascending=True)
```

```python
df
```

```python
fig, axs = plt.subplots(1,2)
sns.set_style("whitegrid")
axs[0].plot('date','suspeitos','orange',data=df)
axs[0].set_title('Suspeitos')
axs[1].plot('date','casos','-r',data=df)
axs[1].set_title('Casos Confirmados')

for ax in axs:
    ax.tick_params(axis='x', rotation=90)
```

* Log Scale Plot

```python
from matplotlib import ticker
g = sns.lineplot(x="date", y="suspeitos",color='orange', data=df)
g = sns.lineplot(x="date", y="casos",color='red', data=df)
g.set(yscale="log")
g.set_ylim([0.5, 1000])
g.legend(['Casos Suspeitos','Casos Confirmados'])
plt.xticks(rotation=90)
plt.show()
```

```python
# sns.set()

ax = sns.lineplot(x="date", y="suspeitos",color='orange', data=df)
ax = sns.lineplot(x="date", y="casos",color='red', data=df)

ax.legend(['Casos Suspeitos','Casos Confirmados'])
ax.set(xlabel='Data', ylabel='Quantidade')

plt.xticks(rotation=90)
plt.show()
```

```python
crDf = df.melt('date',var_name='tipo',value_name='quantidade')
crDf
```

```python
ax = sns.barplot(x='date',y='quantidade',hue='tipo',data=crDf,color='red')
ax.set(xlabel='Data', ylabel='Quantidade')

plt.xticks(rotation=90)
plt.show()
```
